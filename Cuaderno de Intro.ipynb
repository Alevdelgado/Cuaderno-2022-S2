- Mayor o igual a cero

```
	Mientras a>0 Hacer
		a = trunc(a/10)
		total = total + 1

```



````python
def contarDigitos(numero):
    total = 0
    # validar que el numero es entero y >= 0
    if isinstance(numero,int) and (numero >= 0):
        if numero == 0:
            return 1
        else:
            while numero > 0:
                numero = numero // 10
                total += 1
            return total
    else:
        return 'Error: Ingrese un numero entero y positivo'
````

````python
class Cuenta:

    def __init__(self, saldoInicial):
        self.saldoInicial = saldoInicial
        self.montoTotalDebitos = 0
        self.montoTotalCreditos = 0
        self.saldoActual = saldoInicial

    def depositar(self, monto): # credito
        if isinstance(monto, int):
            if monto > 0:
                self.montoTotalCreditos += monto
                self.saldoActual += monto
            else:
                return 'Error: debe ingresar un monto mayor que 0'
        else:
            return 'Error: monto debe ser de tipo INT'
        
    def retirar(self, monto):  # debito
        if isinstance(monto, int):
            if monto > 0 and monto <= self.saldoActual:
                self.montoTotalDebitos += monto
                self.saldoActual -= monto
            else:
                return 'Fondo insuficiente'
        else:
            return 'Error: monto debe ser de tipo INT'

    def determinarEstadoCuenta(self):
        estado = 'Estado de cuenta\n'
        estado += '================\n'
        estado += '\n'
        estado = 'Saldo inicial: ' + str(self.saldoInicial) + ' colones \n'
        estado += 'Total de creditos: ' + str(self.montoTotalCreditos) + ' colones\n'
        estado += 'Total de debitos: ' + str(self.montoTotalDebitos)+ ' colones \n'
        estado += 'Saldo actual: ' + str(self.saldoActual)+ ' colones'
        print(estado)
````

## Circulo - Rectangulo - Calculadora

````python
class Circulo: # definicion del circulo
# toda clase necesita una identidad

# constructor

    def __init__(self, radio):
        self.radio = radio # defino las variables dentro del constructor (instancia)
        self.pi = 3.14

        
    # area de un circulo

    def calcularArea(self):
        resultado = self.pi* (self.radio ** 2)
        return resultado

    # perimetro de un circulo

    def calcularPerimetro(self):
        resultado = 2 * self.pi * self.radio
        return resultado

    def tipo(self):
        return 'es un circulo'




# Rectangulo

class Rectangulo:

    def __init__(self,alto,ancho): # constructor
        self.alto = alto # atributos
        self.ancho = ancho

    def calcularArea(self): # metodos / comportamiento
        resultado = self.alto*self.ancho
        return resultado
    
    def calcularPerimetro(self):
        resultado = (2*self.ancho) + (2*self.alto)
        return resultado

    def verAltura(self):
        altura = self.alto
        return altura

    def verAncho(self):
        ancho = self.ancho
        return ancho


class Calculadora:

    def __init__(self,operador1, operador2):
        self.operador1 = operador1
        self.operador2= operador2

    def suma(self):
        resultado = self.operador1 +self.operador2
        return resultado

    def resta(self):
        resultado = self.operador1 - self.operador2
        return resultado
    
    def multiplicacion(self):
        resultado = self.operador1 * self.operador2
        return resultado

    def division(self):
        if self.operador2 > 0:
            resultado = self.operador1 / self.operador2
            return resultado
        else:
            return 'No se puede dividir entre 0'
        
    def potencia(self):
        resultado = self.operador1 ** self.operador2
        return resultado       
        
````
