## Promedio
Usando Pseint crear un diagrama de flujo para representar el flujo y pseudocodigo para el algoritmo de contarDigitos.

- Recibira cuatro parametros (n1,n2,n3,n4)

### Diagrama de flujos
promedio.png

### Pseudocodigo


```
Algoritmo promedio
	Leer N1
	Leer N2
	Leer N3
	Leer N4
	resultado = (N1+N2+N3+N4)/4
	Escribir resultado
FinAlgoritmo

```


````python
def promedio (n1,n2,n3,n4):
    if (isinstance(n1,int))and(isinstance(n2,int)) and (isinstance(n3, int))and (isinstance(n4,int)):
        while (n1 >= 0 and n2 >= 0 and n3 >= 0 and n4 >= 0 ):
            promedio = (n1+n2+n3+n4)/4
            
            return int(promedio)
    else:
        return 'Error: Ingrese un numero entero'
````

````python
promedio (2,4,6,8)
````
