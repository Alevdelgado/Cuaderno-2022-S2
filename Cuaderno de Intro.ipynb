## Contador de digitos

Usando Pseint crear un diagrama de flujo para representar el flujo y pseudocodigo para el algoritmo de contarDigitos.

- Recibiran un parametro
- Mayor o igual a cero

### Diagrama de flujo 
2022-08-03 (11).png



### Pseudocodigo 



```
Algoritmo contarDigitos
	total = 0
	Leer a
	Mientras a>0 Hacer
		a = trunc(a/10)
		total = total + 1
	FinMientras
	Escribir total
FinAlgoritmo
```



````python
def contarDigitos(numero):
    total = 0
    # validar que el numero es entero y >= 0
    if isinstance(numero,int) and (numero >= 0):
        if numero == 0:
            return 1
        else:
            while numero > 0:
                numero = numero // 10
                total += 1
            return total
    else:
        return 'Error: Ingrese un numero entero y positivo'
````

Usando Pseint crear un diagrama de flujo para representar el flujo y pseudocodigo para el algoritmo de precioTraje.

- Recibira un parametro (precio del traje)
- Los trajes que tienen un precio superior a 2500 se les
aplicar un descuento de 15 %, a todos los dems se les aplicar slo 8 %.

### Diagrama de flujos
precioTraje.png

### Pseudocodigo


```
Algoritmo precioTraje
	Leer precio
	Si precio > 2500 Entonces
		precioFinal = precio -(precio*0.15)
	SiNo
		precioFinal = precio - (precio*0.08)
	FinSi
	Escribir precio
FinAlgoritmo

```



````python
def precioTraje(precio):
  if (isinstance(precio, str)):
    return 'Error: El precio no puede ser de tipo STR'
  else:
    if precio > 0 :
      if precio > 2500:
        precioFinal = precio - (precio*0.15)
        return precioFinal
      else:
        precioFinal = precio - (precio*0.08)
        return precioFinal
    else:
      return 'Error: El precio debe ser mayor a cero, un numero positivo'
````

## Par o Impar

Se requiere un algoritmo dado un nmero si este es par o impar. Como restriccin el nmero debe ser entero y mayor o igual a cero

- esPar(5) False
- esPar(8) True
- esPar(-10) Error: El nmero de ser mayor o igual a CERO




### Diagrama de flujos
esPar.png

### Pseudocodigo


```
Algoritmo esPar
	Leer num
	Si num >= 0 Entonces
		numero = num MOD 2
		Si numero = 0 Entonces
			Escribir 'True'
		SiNo
			Escribir 'False'
		FinSi
	SiNo
		Escribir 'Error: El numero debe ser positivo'
	FinSi
FinAlgoritmo

```



````python
def esPar(num):
  if (isinstance(num,int) and num >= 0):
    numero = num % 2
    if numero == 0:
      return True
    else:
      return False
  else:
    return 'Error: El nmero de ser mayor o igual a CERO'
````

````python
esPar(5)
````

````python
esPar(8)
````

````python
esPar(-10)
````
