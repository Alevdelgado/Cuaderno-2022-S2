## Contador de digitos

Usando Pseint crear un diagrama de flujo para representar el flujo y pseudocodigo para el algoritmo de contarDigitos.

- Recibiran un parametro
- Mayor o igual a cero

### Pseudocodigo 



```

	Mientras a>0 Hacer
		a = trunc(a/10)
		total = total + 1
	FinMientras
	Escribir total
FinAlgoritmo
```



````python
def contarDigitos(numero):
    total = 0
    # validar que el numero es entero y >= 0
    if isinstance(numero,int) and (numero >= 0):
        if numero == 0:
            return 1
        else:
            while numero > 0:
                numero = numero // 10
                total += 1
            return total
    else:
        return 'Error: Ingrese un numero entero y positivo'
````

## Factorial 
Nombre: factorial()
Entradas: Se requiere un algoritmo para obtener el factorial desde 1 a un número cualquiera. 
Salidas: Un valor que indica el resultado del factorial de un numero
Restricciones:  Debe ser entero y no negativo
factorial(5)

factorial(3)

factorial(-5)
“Error: El número de ser mayor a CERO”

### Diagrama de flujos 

factorial.png



### Pseudocodigo


```
Algoritmo factorial
	resultado <- 1
	i <- 2
	Leer num
	Si num>=0 Entonces
		Mientras i<=num Hacer
			resultado <- resultado*i
			i <- i+1
		FinMientras
		i = i + 0

		Escribir 'Error: El numero debe ser entero positivo'
```

````python
def factorial(num):
    resultado = 1
    i = 2 # contador 
    if (isinstance(num, int) and num >= 0):
        if num == 0:
            return 1
        while i <= num :
            resultado *= i
            i += 1
        return resultado
            
    else:
        return 'Error: El numero debe ser mayor a CERO'
````

## Es numero primo
Construir una funcionque reciba un numero e indique si es primo.

- Recibe un parametro (numero entero positivo)

### Diagrama de flujos
esPrimo.png

### Pseudocodigo


```
Algoritmo esPrimo
	Leer num
	Si num > 0 Entonces
		Si num = 1 Entonces
			Escribir FALSO
		SiNo
			Si num = 2 Entonces
				Escribir VERDADERO
			SiNo
				divisor = 2
				Mientras divisor<= num  Hacer
					Si num MOD 1 = 0 y num MOD divisor = 0  Entonces
						Escribir FALSO
					SiNo
						Escribir VERDADERO
					FinSi
					divisor = divisor + 1
				FinMientras
			FinSi
		FinSi
	SiNo
		Escribir 'Error: numero no valido'
	FinSi
FinAlgoritmo

```



````python
def esPrimo(num):
    divisor = 2
    if (isinstance(num, int) and num > 0):
        if num == 1:
            return False
        if num == 2:
            return True
        while divisor <= num:          
            if num % 1 == 0 and num % divisor == 0:
                return False
            else:
                return True               
        divisor += 1
                         
    else:
        return 'Error: en la entrada de los parametros'
````
